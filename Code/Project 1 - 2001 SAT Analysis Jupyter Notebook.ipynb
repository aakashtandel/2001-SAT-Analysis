##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def std_dev(dictionary_key):  # We begin by creating a function that takes a list (which in our case would pass a key from the dictionary).
    mu = sum(dictionary_key)/len(dictionary_key)  # We calculate the mean of the list. 
    diffsq = [(each-mu)**2 for each in dictionary_key]  # We generate another list of the difference from the mean squared. 
    std_dev = (sum(diffsq) * 1/len(diffsq))**0.5  # We sum that list, multiply it by 1/n, and square root it to get the standard deviation. 
    return std_dev
```

```python
std_dev(dictionary_data['Rate'])
```

```python
std_dev(dictionary_data['Verbal'])
```

```python
std_dev(dictionary_data['Math'])
```

But this data is the average score of the SAT tests per state. That may not be normally distributed. If we had a larger number of states (n=5,000 states) we may see normality in the data. 

##### 18. Does that distribution hold true for our data?

The average math score data looks close to normally distributed but I would not claim it is so without more data.
